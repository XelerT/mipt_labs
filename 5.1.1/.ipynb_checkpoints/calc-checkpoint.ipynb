```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

w1 = 1
c = 1

# Определение функций
def x(tau):
  return w1*c*(tau**2)/2

def y(tau):
  return -c*w1*w1*(tau**3)/6

def t(tau):
  return w1**2*(tau**3)/6 + tau

# Создание данных
tau_values = np.linspace(0, 10, 100)
t_values = t(tau_values)
x_values = x(tau_values)
y_values = y(tau_values)

# Создание 3D графика
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_box_aspect([1, 1, 1])
ax.plot(t_values, x_values, y_values) 
ax.set_xlabel('t')
ax.set_ylabel('X')
ax.set_zlabel('Y')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

w1 = 1
w2 = w1/2
c = 1

# Определение функций
def x(tau):
  return np.real(c*w1**2/(w1**2 - w2**2)*np.cosh(np.sqrt(complex(w1**2 - w2**2))*tau))

def y(tau):
  return np.real(c*w1*w2/(w1**2 - w2**2)*(tau + np.sinh(np.sqrt(complex(w1**2 - w2**2))*tau)))

def t(tau):
  return np.real(w1**2/(w1**2 - w2**2)**(3/2)*np.sinh(np.sqrt(complex(w1**2 - w2**2))*tau) - w2**2/(w1**2 - w2**2)*tau)

# Создание данных
tau_values = np.linspace(0, 10, 100)
t_values = t(tau_values)
x_values = x(tau_values)
y_values = y(tau_values)

# Создание 3D графика
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_box_aspect([1, 1, 1])
ax.plot(t_values, x_values, y_values) 
ax.set_xlabel('t')
ax.set_ylabel('X')
ax.set_zlabel('Y')
plt.show()
```
