```python
class consts:
    pi = np.pi
    k_b = 1.38e-23
    c = 3e8

class setup:
    dummy = 0
    T_k = 298 # K
    P_k = 1.03 * 760 # tor
    geiger_time = 20 # s
    R_0 = 9 # cm
    T_0 = 273
    R_0_ion = 5 # cm
    
```

## Счетчик Гейгера

```python
data = mipt.read_excel('data/5.4.1.xlsx', usecols=(0,1,2), header=(0)).dropna()
data
```

```python
from scipy.interpolate import interp1d

l = data['l']
N = data['N'] / data['t']
dN_dl = np.gradient(N, l)

coefficients = np.polyfit(l, N, 4)
polynomial = np.poly1d(coefficients)
deriv = polynomial.deriv()

def get_N_l_polynom(l, N):
    coefficients = np.polyfit(l, N, 10)
    return np.poly1d(coefficients)

mipt.plot(l, N, func=get_N_l_polynom, marker_size=7, exclude=[], linestyle='dashed')
x, y, equ = mipt.plot(np.array(l[6:11]), np.array(N[6:11]), marker_size=0, color="red")

plt.axhline(y=0, color='black', linestyle='--', linewidth=0.8, label="y = 0")
plt.axvline(x=0, color='black', linestyle='--', linewidth=0.8, label="y = 0")
plt.xlabel('l, $\\text{мм}$', fontsize=10, rotation=0)
plt.ylabel('$N$, $c^{-1}$', fontsize=10)


plt.plot(l[3:14], -3*deriv(l[3:14]), linestyle='dashed')

mnk = mipt.mnk(x, y)
mnk.to_latex('gen/tab-mnk-geig.tex')
print(mnk.data)

plt.savefig("gen/geig.pdf")
```

```python
l_fine = np.linspace(l.min(), l.max(), 1000)
dN_dl_fine = -3*deriv(l_fine)

max_index = np.argmax(dN_dl_fine)

l_av = l_fine[max_index]
l_av_err = l_av * 0.12 # No time to think
print("R_ср=", l_av, " мм")
R_ef = (-mnk.data['b'] / mnk.data['a'])[0]
R_ef_err = R_ef * np.sqrt((mnk.data['da'][0] / mnk.data['a'][0])**2 + (mnk.data['db'][0] / mnk.data['b'][0])**2)
print("R_э=", R_ef,"+-", R_ef_err, " мм")

E_ef = (R_ef/0.32)**(2/3)
E_ef_err = R_ef_err / R_ef * E_ef
print("E_eff=", E_ef, "+-", E_ef_err, "Мэв")

E_av = (l_av/0.32)**(2/3)
E_av_err = l_av_err / l_av * E_av
print("E_ср=", E_av, "+-", E_av_err, "Мэв")
```

## Сцинциляционный счетчк

```python
data = mipt.read_excel('data/5.4.1.xlsx', usecols=(3,4,5,7), header=(0)).dropna()
data

```

```python
P = setup.P_k - data['p']
N = (data['N1'] + data['N2'] + data['N3']) / 3

coefficients = np.polyfit(P, N, 5)
polynomial = np.poly1d(coefficients)
deriv = polynomial.deriv()

def get_N_P_polynom(P, N):
    coefficients = np.polyfit(P, N, 8)
    return np.poly1d(coefficients)

mipt.plot(P, N, func=get_N_P_polynom, marker_size=7, exclude=[])
plt.plot(P[0:6], -120*deriv(P[0:6]), linestyle='dashed')

x, y, equ = mipt.plot(np.array(P[:6]), np.array(N[:6]), marker_size=0, color="red")

plt.axhline(y=0, color='black', linestyle='--', linewidth=0.8, label="y = 0")
plt.axvline(x=0, color='black', linestyle='--', linewidth=0.8, label="y = 0")
plt.xlabel('P, мм.рт.ст', fontsize=10)
plt.ylabel('$N$', fontsize=10)

mnk = mipt.mnk(x, y)
mnk.to_latex('gen/tab-mnk-scinc.tex')
print(mnk.data)

plt.savefig("gen/scin.pdf")
```

```python
P_fine = np.linspace(P.min(), P.max(), 1000)
dN_dP_fine = -120*deriv(P_fine)

max_index = np.argmax(dN_dP_fine)

P_max = P_fine[max_index]
print(P_max)

R_max = setup.R_0 * P_max * setup.T_k / (setup.P_k* setup.T_0)
print("R_ср=", R_max, " мм")

P_ef = (-mnk.data['b'] / mnk.data['a'])[0]
R_ef = setup.R_0 * P_ef * setup.T_k / (setup.P_k* setup.T_0)
R_ef_err = R_ef * np.sqrt((mnk.data['da'][0] / mnk.data['a'][0])**2 + (mnk.data['db'][0] / mnk.data['b'][0])**2)
print(P_ef)

print("R_э=", R_ef,"+-", R_ef_err, " cм")

E = (R_ef/0.32)**(2/3)
E_err = R_ef_err / R_ef * E
print(E, "+-", E_err, "Мэв")
```

## Ионизационная камера

```python
data = mipt.read_excel('data/5.4.1.xlsx', usecols=(6,7), header=(0)).dropna()
data
```

```python
P = setup.P_k - data['p']
I = data['I']

inc = np.polyfit(np.array(P[:-3]), np.array(I[:-3]), deg=1)
x = np.linspace(min(P), max(P), 10000)
y =  inc[0] * x + inc[1]
plt.plot(x, y,"-.", color='red')

dec = np.polyfit(np.array(P[-3:]), np.array(I[-3:]), deg=1)

x_dec, y_dec, equ = mipt.plot(np.array(P[:-3]), np.array(I[:-3]), marker_size=4, color="red")
x, y, equ = mipt.plot(np.array(P[-3:]), np.array(I[-3:]), marker_size=4, color="blue")
plt.axhline(y=0, color='black', linestyle='--', linewidth=0.8, label="y = 0")
plt.axvline(x=0, color='black', linestyle='--', linewidth=0.8, label="y = 0")
plt.xlabel('P, мм.рт.ст', fontsize=10)
plt.ylabel('I, пА', fontsize=10)

plt.savefig("gen/ion.pdf")
```

```python
P_eq = (dec[1] - inc[1]) / (inc[0] - dec[0])
mnk_dec = mipt.mnk(x_dec, y_dec)
mnk_dec.to_latex('gen/tab-mnk-dec.tex')
mnk_inc = mipt.mnk(x, y)
mnk_inc.to_latex('gen/tab-mnk-inc.tex')
print(mnk_dec.data)
print()
print(mnk_inc.data)
P_eq_err = error = P_eq * np.sqrt((mnk_dec.data['da'] / mnk_dec.data['a'])**2 + (mnk_dec.data['db'] / mnk_dec.data['b'])**2)[0]

print("P_0=", P_eq, "+-", P_eq_err, "Tor")

R_norm = setup.R_0 * P_eq * setup.T_k / (setup.P_k* setup.T_0)
R_norm_err = R_norm * P_eq_err/P_eq
print("R_norm=", R_norm, "+-", R_norm_err, "cm")

E_norm = (R_norm/0.32)**(2/3)
E_norm_err = R_norm_err / R_norm * E_norm
print("E_norm=", E_norm, "+-", E_norm_err, "Мэв")
```
